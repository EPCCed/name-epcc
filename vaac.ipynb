```python
# VAAC
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Thread efficiency and task efficiency (from speedup)
n_groups = 3

eUpdate = (1.0, 9.7/12, 16.0/24)
eOutput = (1.0, 2.0/12.0, 2.0/24)

fig, (ax1, ax2) = plt.subplots(1,2)
fig.set_figwidth(10.0)
#fig.subplots_adjust(left=0.2, wspace=0.6)

index = np.arange(n_groups)
bar_width = 0.30
opacity = 0.4

rects1 = ax1.bar(index, eUpdate, bar_width,
                alpha=opacity, color='b', label='Update')
rects2 = ax1.bar(index + bar_width, eOutput, bar_width,
                alpha=opacity, color='r', label='Output')

ax1.set_xlabel('Threads', fontsize = 16)
ax1.set_ylabel('Parallel Efficiency', fontsize = 16)
ax1.set_title(' Smaller', fontsize = 16, loc = "left")
ax1.set_xticks(index + bar_width / 2)
ax1.set_xticklabels(('1', '12', '24'), fontsize = 16)
ax1.legend(fontsize = 16)

# MPI for extended banchmarks 1 and 2

e1update = (1.0, 2.0/2.0, 3.8/4.0)
e2output = (1.0, 0.8/2.0, 0.7/4.0)
rects1 = ax2.bar(index, e1update, bar_width,
                alpha=opacity, color='b', label='Update')
rects2 = ax2.bar(index + bar_width, e2output, bar_width,
                alpha=opacity, color='r', label='Output')

ax2.set_xlabel('MPI Tasks', fontsize = 16)
ax2.set_ylabel('Parallel Efficiency', fontsize = 16)
ax2.set_title(' Larger', fontsize = 16, loc = 'left')
ax2.set_xticks(index + bar_width / 2)
ax2.set_xticklabels(('1', '2', '4'), fontsize = 16)
#ax2.legend(fontsize = 16)
fig.tight_layout()
plt.savefig("BenchVaac.svg")
plt.show()
```
